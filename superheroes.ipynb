## Retrieve total of first 5 records of max power of good alignment superheroes
Use nlarges and pass arguments (number of rows to return, columns label for the output to be ordered by)

```python
df_res=df[['Name','Total','Power']].loc[ (df['Alignment']=='good') ].nlargest(5, 'Power')
df_res
```

## Draw bar plot of all superheroes with good alignment and max power of top 5, use object of prompt 13
Use df.sort_values to grab the dataframe of superheroes with good alignment, max power and top 5 highest total power. Then use plot.bar to draw the bar plot.

```python
df_res=df[['Name','Total','Power']].loc[ (df['Alignment']=='good') ].sort_values(by=['Total'],ascending=False).head(5)
power = df_res['Total']
superhero_name = df_res['Name']
plt.bar(superhero_name,power, color = 'green')
plt.xticks(rotation = 90)
plt.title("Power of superheroes with max power")
plt.xlabel("Name")
plt.ylabel("Total Power level")
plt.show()
```
